```python
#DATA ENCODING
df['Gender'] = df['Gender'].astype('category')
print(df.dtypes)
```

```python
df
```

```python
Gender_dummies = pd.get_dummies(df['Gender'], prefix = 'Gender' )
df = pd.concat([df,Gender_dummies], axis = 1)
print(df)
```

```python
df['Branch'] = df['Branch'].astype('category')
df['City'] = df['City'].astype('category')
df['Customer type'] = df['Customer type'].astype('category') 
print(df.dtypes)
```

```python
Branch_dummies = pd.get_dummies(df['Branch'], prefix = 'Branch' )
df = pd.concat([df,Branch_dummies], axis = 1)
print(df)
```

```python
City_dummies = pd.get_dummies(df['City'], prefix ="City")
df = pd.concat([df,City_dummies], axis = 1)
print(df)
```

```python
CustomerType_dummies = pd.get_dummies(df['Customer type'], prefix = 'Customer type' )
df = pd.concat([df,CustomerType_dummies], axis = 1)
print(df)
```

```python
#DATA VISUALIZATION
df.plot(kind = 'bar')
```

```python
df.drop('City', axis = 1)
df.plot(kind = 'bar')
```
